## Map, Apply, ApplyMap

### Map

```python
data = {'first_name' : ['Meena','Sudhir','Kumar'], 'age' : [23, 48, 16], 'city': ['Mumbai','Boston','Sydney']}
df_x = pandas.DataFrame(data, columns=['first_name','age','city'])
df_x
```

```python
city_to_country = {'Mumbai':'India', 'Boston':'USA', 'Sydney':'Australia'}
city_to_country
```

```python
df_x['country'] = df_x['city'].map(city_to_country)
df_x
```

### Apply

```python
capitalizer = lambda x: x.upper()
df_x["first_name"] = df_x["first_name"].apply(capitalizer)
df_x
```

### ApplyMap

```python
data_y = {'height':[74, 52, 5, 27], 'age':[81, 64, 4, 25]}
df_y = pandas.DataFrame(data_y, columns=['height','age'])
df_y
```

```python
df_y.applymap(np.sqrt)
```
