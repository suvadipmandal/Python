## import pandas package

```python
import pandas as pd
import numpy as np
```

```python
HR_data = pd.read_csv('5000 Records.csv')
columns_name = HR_data.columns.tolist()
print("Columns name :")
print(columns_name)
HR_data.head()
```

```python
women=HR_data[HR_data.Gender=='M']
t_male=women.shape
t_emp=HR_data.shape
print("Total Male Employee's="+str(t_male[0]))
print("Total Female Employee's="+str(t_emp[0]-t_male[0]))
```

```python
HR_data=HR_data.rename(columns={'Emp ID': 'Emp_ID',
                                'Name Prefix':'Name_Prefix',
                                'First Name':'F_Name',
                                'Middle Initial':'M_Name',
                                'Last Name':'L_Name',
                                'E Mail':'EMail',
                                'Father\'s Name':'F_Name',
                                'Mother\'s Name':'M_Name',
                                'Mother\'s Maiden Name':'M_M_Name',
                                'Date of Birth':'DOB',
                                'Time of Birth':'TOB',
                                'Age in Yrs.':'AIY',
                                'Weight in Kgs.':'Weight',
                                'Date of Joining':'DOJ',
                                'Half of Joining':'HOJ',
                                'Quarter of Joining':'QOJ',
                                'Year of Joining':'YOJ',
                                'Month of Joining':'MOJ',
                                'Month Name of Joining':'MNOJ',
                                'Short Month':'SM',
                                'Day of Joining':'Day_Of_Joining',
                                'DOW of Joining':'DOW_of_Joining',
                                'Short DOW':'S_DOW',
                                'Age in Company (Years)':'time_spend_company',
                                'Last % Hike':'Increment',
                                'Phone No.':'P_No',
                                'Place Name':'Place_Name',
                                'User Name':'User_Name'})
```

```python
HR_data.columns
```

```python
t_data=HR_data['time_spend_company'].unique()
t_data.sort()
t_data.tolist()
```

```python

HR_data['Departments'] = HR_data['time_spend_company'].apply(lambda x: 'Tranning' if (x >= 0.00) & (x <= 0.09)
                                                             else 'Internship' if (x >= 0.1) & (x <= 0.99)  
                                                             else 'Sales' if (x >= 1.0) & (x <= 6.0)
                                                             else 'Markating' if (x >= 6.01) & (x <= 10.0)
                                                             else 'Technical' if (x >= 10.01) & (x <= 11.0)
                                                             else 'product_mng' if (x >= 11.01) & (x <= 11.2)
                                                             else 'Management')
HR_data.loc[(HR_data['time_spend_company'] >= 11.22) & (HR_data['Gender']=='F') & (HR_data['Salary']>199000), 'Departments'] = 'HR'
HR_data.loc[(HR_data['time_spend_company'] >= 11.26) & (HR_data['Salary']>199950), 'Departments'] = 'CEO'
```

```python
HR_data[HR_data['Departments']=="HR"]
```

```python
a=HR_data['Departments'].isnull()
count=0
for x in range(len(a)):
    if a[x]==True:
        count+=1

print(count)
HR_data.head()
```
